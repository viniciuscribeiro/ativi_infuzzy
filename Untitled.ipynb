```python
%matplotlib inline
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


qualidade_equipamentos = ctrl.Antecedent(np.arange(0, 11, 1), 'qualidade_equipamentos')
ergonomia = ctrl.Antecedent(np.arange(0, 11, 1), 'ergonomia')
iluminacao = ctrl.Antecedent(np.arange(0, 11, 1), 'iluminacao')  
temperatura = ctrl.Antecedent(np.arange(0, 11, 1), 'temperatura')  
avaliacao_final = ctrl.Consequent(np.arange(0, 11, 1), 'avaliacao_final')

qualidade_equipamentos['ruim'] = fuzz.trimf(qualidade_equipamentos.universe, [0, 0, 5])
qualidade_equipamentos['medio'] = fuzz.trimf(qualidade_equipamentos.universe, [0, 5, 10])
qualidade_equipamentos['bom'] = fuzz.trimf(qualidade_equipamentos.universe, [5, 10, 10])

ergonomia['ruim'] = fuzz.trimf(ergonomia.universe, [0, 0, 5])
ergonomia['medio'] = fuzz.trimf(ergonomia.universe, [0, 5, 10])
ergonomia['bom'] = fuzz.trimf(ergonomia.universe, [5, 10, 10])

iluminacao['ruim'] = fuzz.trimf(iluminacao.universe, [0, 0, 5])
iluminacao['medio'] = fuzz.trimf(iluminacao.universe, [0, 5, 10])
iluminacao['bom'] = fuzz.trimf(iluminacao.universe, [5, 10, 10])

temperatura['frio'] = fuzz.trimf(temperatura.universe, [0, 0, 5])
temperatura['medio'] = fuzz.trimf(temperatura.universe, [0, 5, 10])
temperatura['quente'] = fuzz.trimf(temperatura.universe, [5, 10, 10])

avaliacao_final['ruim'] = fuzz.trimf(avaliacao_final.universe, [0, 0, 5])
avaliacao_final['medio'] = fuzz.trimf(avaliacao_final.universe, [0, 5, 10])
avaliacao_final['bom'] = fuzz.trimf(avaliacao_final.universe, [5, 10, 10])

rule1 = ctrl.Rule(qualidade_equipamentos['bom'] & ergonomia['bom'] & temperatura['medio'], avaliacao_final['bom'])
rule2 = ctrl.Rule(qualidade_equipamentos['medio'] & iluminacao['ruim'] & temperatura['quente'], avaliacao_final['medio'])
rule3 = ctrl.Rule(qualidade_equipamentos['ruim'] | ergonomia['ruim'] | temperatura['frio'], avaliacao_final['ruim'])
rule4 = ctrl.Rule(iluminacao['bom'] & temperatura['medio'] & qualidade_equipamentos['medio'], avaliacao_final['medio'])  

sistema_controle = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
sistema_simulacao = ctrl.ControlSystemSimulation(sistema_controle)


sistema_simulacao.input['qualidade_equipamentos'] = 7.5
sistema_simulacao.input['ergonomia'] = 6.5
sistema_simulacao.input['iluminacao'] = 4.0  
sistema_simulacao.input['temperatura'] = 5.0  

sistema_simulacao.compute()

print(f"Avaliação Final: {sistema_simulacao.output['avaliacao_final']}")

avaliacao_final.view(sim=sistema_simulacao)
```
